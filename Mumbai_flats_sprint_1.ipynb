```python
import pandas as pd
flats_df=pd.read_csv("flats.csv")
flats_df.head()
## dropping link and propertyid
flats_df.drop(columns={'link','property_id'},axis=0,inplace=True )
flats_df
flats_df.head()
```

```python
## extracting type of BHK from property_name
flats_df['property_name'].unique()
flats_df['property_type']=flats_df['property_name'].apply(lambda x:x.split(" ")[0:2])
flats_df['property_type']=flats_df['property_type'].apply(lambda x: str(x).replace("'",""))
flats_df['property_type']=flats_df['property_type'].apply(lambda x: str(x).replace("[",""))
flats_df['property_type']=flats_df['property_type'].apply(lambda x: str(x).replace("]",""))
flats_df['property_type']=flats_df['property_type'].apply(lambda x: str(x).replace(", "," "))
##rearranging column positions
flats_df=flats_df[['property_name','property_type','society','price','area','areaWithType','bedRoom','bathroom','balcony','additionalRoom',
'address','floorNum','facing','agePossession','nearbyLocations','description','furnishDetails','features','rating']]
flats_df.head(3)
```

```python
flats_df[flats_df['price'].isnull()]
flats_df.drop(index=18,axis=0,inplace=True)
```

```python
flats_df['value'] = flats_df['price'].apply(lambda x: str(x).split()[0])
flats_df['currency'] = flats_df['price'].apply(lambda x: str(x).split()[1:2])
flats_df['currency'] = flats_df['currency'].apply(lambda x: str(x).replace("[", ""))
flats_df['currency'] = flats_df['currency'].apply(lambda x: str(x).replace("]", ""))
flats_df['currency'] = flats_df['currency'].apply(lambda x: str(x).replace("'", ""))
```

```python
flats_df=flats_df[~(flats_df['price']=='Price on Request')]
flats_df=flats_df[~(flats_df['price']=='Price on Request')]
```

```python
flats_df[flats_df['price']=="Price on Request "]
```

```python
import re
flats_df['society'] = flats_df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
flats_df['price']=flats_df['price'].str.replace("'","")
flats_df['price']=flats_df['price'].str.replace("'","")
flats_df['price'].head(26)
```

```python

def get_price(x):
  if x.split()[1]=="Lac":
    return round(float(x.split()[0])/100,2)
  else:
    return float(x.split()[0])

flats_df['price']=flats_df['price'].apply(get_price)




```

```python
flats_df.head()
# rename columns
flats_df.rename(columns={'area':'price_per_sqft'},inplace=True)
flats_df.head()
```

```python
flats_df['price_per_sqft'].value_counts()
```

```python
flats_df['price_per_sqft'] = flats_df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype(float)
```

```python
#flats_df['price']=flats_df['price'].astype(float)
flats_df.info()
```

```python
# bedrooms
flats_df['bedRoom'].value_counts()
```

```python
flats_df[flats_df['bedRoom'].isnull()]
flats_df['bedRoom'] = flats_df['bedRoom'].str.split(' ').str.get(0).astype('int')
```

```python
# bathroom
flats_df['bathroom'].value_counts()
flats_df['bathroom'] = flats_df['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
# balcony
flats_df['balcony'].value_counts()
flats_df['balcony'] = flats_df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
flats_df['balcony'].value_counts()
```

```python
# additionalRoom
flats_df['additionalRoom'].value_counts()
```

```python
flats_df['additionalRoom'].fillna('not available',inplace=True)
flats_df['additionalRoom'] = flats_df['additionalRoom'].str.lower()
flats_df.shape[0]
```

```python
flats_df['floorNum']=flats_df['floorNum'].str.strip(" ")
```

```python
flats_df[~flats_df['floorNum'].isnull()]
```

```python
flats_df['floor_Num'] = flats_df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
flats_df['total_floors']=flats_df['floorNum'].str.extract(r'of (\d+)')
```

```python
flats_df.insert(loc=4,column='area',value=round((flats_df['price']*10000000)/flats_df['price_per_sqft']))
```
